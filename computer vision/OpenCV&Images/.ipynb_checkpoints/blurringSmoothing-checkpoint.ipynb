```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_img():
    img = cv2.imread("../mmexport1545548449041.jpg").astype(np.float32) / 255
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    return img
```

```python
# load_img()
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img)
```

```python
i = load_img()
display_img(i)
```

```python
# Gamma correction
gamma = 4
```

```python
result = np.power(i, gamma)
display_img(result)
```

```python
# Blurring
img = load_img()
font = cv2.FONT_HERSHEY_COMPLEX
cv2.putText(img, text="Ivan", org=(10,600),fontFace=font, fontScale=10,color=(255,0,0),thickness=4)
```

```python
display_img(img)
```

```python
kernel = np.ones(shape=(5,5),dtype=np.float32)/25
```

```python
kernel
```

```python
dst = cv2.filter2D(img, -1, kernel)
```

```python
display_img(dst)
```

```python
# Blurring
img = load_img()
font = cv2.FONT_HERSHEY_COMPLEX
cv2.putText(img, text="Ivan", org=(10,600),fontFace=font, fontScale=10,color=(255,0,0),thickness=4)
print("reset")
```

```python
blurred = cv2.blur(img, ksize=(20,20))
display_img(blurred)
```

```python
# Blurring
img = load_img()
font = cv2.FONT_HERSHEY_COMPLEX
cv2.putText(img, text="Ivan", org=(10,600),fontFace=font, fontScale=10,color=(255,0,0),thickness=4)
print("reset")
```

```python
blurred = cv2.GaussianBlur(img, (5,5), 10)
display_img(blurred)
```

```python
# Median Blurring - removing image noises
img = load_img()
font = cv2.FONT_HERSHEY_COMPLEX
cv2.putText(img, text="Ivan", org=(10,600),fontFace=font, fontScale=10,color=(255,0,0),thickness=4)
print("reset")
```

```python
blurred = cv2.medianBlur(img, 5)
display_img(blurred)
```
