```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_image():
    blank_img = np.zeros((600,600))
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(blank_img, text = 'ABCDE', org = (50,300), fontFace=font, fontScale=5, color=(255,255,255),thickness=25)
    return blank_img
```

```python
def display_img(img):
    fig = plt.figure(figsize = (12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
img = load_image()
```

```python
display_img(img)
```

```python
kernel = np.ones((5,5), dtype = np.uint8)
kernel
```

```python
result = cv2.erode(img, kernel, iterations =5)
```

```python
display_img(result)
```

```python
img = load_image()
```

```python
white_noise = np.random.randint(low=0, high=2,size=(600,600))
```

```python
display_img(white_noise)
```

```python
white_noise = white_noise*255
```

```python
display_img(white_noise)
```

```python
noise_img = white_noise+img
display_img(noise_img)
```

```python
opening = cv2.morphologyEx(noise_img, cv2.MORPH_OPEN, kernel)
```

```python
display_img(opening)
```

```python
black_noise = np.random.randint(0,2,(600,600))
black_noise = black_noise * -255
```

```python
display_img(black_noise)
```

```python
black_noise_image = img + black_noise
```

```python
black_noise_image[black_noise_image == -255] = 0
display_img(black_noise_image)
```

```python
closing = cv2.morphologyEx(black_noise_image, cv2.MORPH_CLOSE, kernel)
```

```python
display_img(closing)
```
