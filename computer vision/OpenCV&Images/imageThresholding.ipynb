```python
img = cv2.imread("../mmexport1545548449041.jpg", 0)
```

```python
def show_pic(img):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap="gray")
```

```python
show_pic(img)
```

```python
ret, th1 = cv2.threshold(img, 180, 255, cv2.THRESH_BINARY)
```

```python
show_pic(th1)
```

```python
th2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY,15,8)
show_pic(th2)
```

```python
blended = cv2.addWeighted(src1 = th1, alpha = 0.8, src2 = th2, beta = 0.2, gamma = 0.5)
```

```python
show_pic(blended)
```
