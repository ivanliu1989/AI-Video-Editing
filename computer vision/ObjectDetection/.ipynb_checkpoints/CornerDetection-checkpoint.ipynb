```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
full = cv2.imread('../mmexport1545548449041.jpg')
full = cv2.cvtColor(full, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(full)
```

```python
gray_full = cv2.cvtColor(full, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray_full, cmap = 'gray')
```

```python
gray = np.float32(gray_full)
```

```python
dst = cv2.cornerHarris(gray, blockSize=2, ksize=3, k=0.04)
```

```python
dst = cv2.dilate(dst, None)
```

```python
dst
```

```python
full[dst>0.01*dst.max()] = [255,0,0]
```

```python
plt.imshow(full)
```

```python
# Shi-Tomasi Detection
full = cv2.imread('../mmexport1545548449041.jpg')
full = cv2.cvtColor(full, cv2.COLOR_BGR2RGB)
gray_full = cv2.cvtColor(full, cv2.COLOR_BGR2GRAY)
```

```python
corners = cv2.goodFeaturesToTrack(gray_full, 800, 0.01, 10)
```

```python
corners = np.int0(corners)
```

```python
for i in corners:
    x,y = i.ravel()
    cv2.circle(full, (x,y), 3, (255,0,0),-1)
```

```python
plt.imshow(full)
```
