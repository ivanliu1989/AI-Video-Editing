```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def display(img, cmap = 'gray'):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap = cmap)
```

```python
ivan1 = cv2.imread('../mmexport1545548449041.jpg')
ivan1 = cv2.cvtColor(ivan1, cv2.COLOR_BGR2RGB)
```

```python
display(ivan1)
```

```python
ivan2 = cv2.imread('../mmexport1545548420761.jpg')
ivan2 = cv2.cvtColor(ivan2, cv2.COLOR_BGR2RGB)
display(ivan2)
```

```python
ivan_face= cv2.imread('../ivan_face.jpg')
ivan_face = cv2.cvtColor(ivan_face, cv2.COLOR_BGR2RGB)
display(ivan_face)
```

```python
orb = cv2.ORB_create()
```

```python
kp1, des1 = orb.detectAndCompute(ivan1, None)
kp2, des2 = orb.detectAndCompute(ivan2, None)
```

```python
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck = True)
```

```python
matches = bf.match(des1, des2)
```

```python
single_match = matches[0]
```

```python
single_match.distance
```

```python
matches = sorted(matches, key=lambda x:x.distance)
```

```python
ivan_matches = cv2.drawMatches(ivan1, kp1, ivan2, kp2, matches[:50], None, flags = 2)
```

```python
display(ivan_matches)
```

```python
### SIFT
sift = cv2.xfeatures2d.SIFT_create()
```

```python
kp1, des1 = sift.detectAndCompute(ivan1, None)
kp2, des2 = sift.detectAndCompute(ivan2, None)
```

```python
bf = cv2.BFMatcher()
```

```python
matches = bf.knnMatch(des1, des2, k=2)
```

```python
good = []

for match1, match2 in matches:
    if match1.distance < 0.75* match2.distance:
        good.append([match1])
```

```python
len(good)
```

```python
len(matches)
```

```python
sift_matches = cv2.drawMatchesKnn(ivan1, kp1, ivan2, kp2, good, None, flags = 2)
```

```python
display(sift_matches)
```

```python
### FLANN based Matcher
sift = cv2.xfeatures2d.SIFT_create()
kp1, des1 = sift.detectAndCompute(ivan1, None)
kp2, des2 = sift.detectAndCompute(ivan2, None)
```

```python
FLANN_INDEX_KDTREE = 0
index_params = dict(algorithm=FLANN_INDEX_KDTREE, trees = 5)
search_params = dict(checks=50)
```

```python
flann = cv2.FlannBasedMatcher(index_params, search_params)
```

```python
matches = flann.knnMatch(des1, des2, k=2)
```

```python
good = []

for match1, match2 in matches:
    if match1.distance < 0.75* match2.distance:
        good.append([match1])
```

```python
len(good)
```

```python
flann_matches = cv2.drawMatchesKnn(ivan1, kp1, ivan2, kp2, good, None, flags = 0)
```

```python
display(flann_matches)
```
