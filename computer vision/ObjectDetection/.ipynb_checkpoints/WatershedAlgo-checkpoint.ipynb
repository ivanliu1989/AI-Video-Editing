```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def display(img, cmap = 'gray'):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap = cmap)
```

```python
full = cv2.imread('../mmexport1545548449041.jpg')
display(full)
```

```python
# Median Blur
# Grayscale
# Binary Threshold
# Find Contours
```

```python
sep_blur = cv2.medianBlur(full, 25)
```

```python
display(sep_blur)
```

```python
gray_full = cv2.cvtColor(full, cv2.COLOR_BGR2GRAY)
```

```python
display(gray_full)
```

```python
ret, sep_thresh = cv2.threshold(gray_full, 160, 255, cv2.THRESH_BINARY_INV)
```

```python
display(sep_thresh)
```

```python
image, contours, hie = cv2.findContours(sep_thresh.copy(), cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)
```

```python
for i in range(len(contours)):
    
    if hie[0][i][3] == -1:
        cv2.drawContours(full, contours, i, (255,0,0),10)
```

```python
display(full)
```

```python
img = cv2.imread('../mmexport1545548449041.jpg')
img = cv2.medianBlur(img, 35)
```

```python
display(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
ret, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY_INV)
```

```python
display(thresh)
```

```python
# Noise removal
```

```python
kernel = np.ones((3,3), np.uint8)
```

```python
opening = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel, iterations = 2)
```

```python
display(opening)
```

```python
# distance transform
dist_trans = cv2.distanceTransform(opening, cv2.DIST_L2,5 )
```

```python
display(dist_trans)
```

```python
ret, sure_fg = cv2.threshold(dist_trans, 0.7*dist_trans.max(),255,0)
```

```python
display(sure_fg)
```

```python
sure_fg = np.uint8(sure_fg)
```

```python
unknown = cv2.subtract(opening, sure_fg)
```

```python
display(unknown)
```

```python
ret, markers = cv2.connectedComponents(sure_fg)
```

```python
markers  =markers + 1
```

```python
markers[unknown==255]=0
```

```python
display(markers)
```

```python
markers = cv2.watershed(img, markers)
```

```python
display(markers)
```
