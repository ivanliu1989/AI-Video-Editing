```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
full = cv2.imread('../mmexport1545548449041.jpg')
full = cv2.cvtColor(full, cv2.COLOR_BGR2RGB)
```

```python
edges = cv2.Canny(full, threshold1=127, threshold2=127)
```

```python
plt.imshow(edges)
```

```python
edges = cv2.Canny(full, threshold1=225, threshold2=225)
plt.imshow(edges)
```

```python
med_val = np.median(full)
lower = int(max(0, 0.7*med_val))
upper = int(min(255, 1.3*med_val))
```

```python
edges = cv2.Canny(full, threshold1=lower, threshold2=upper)
plt.imshow(edges)
```

```python
blur_full = cv2.blur(full, ksize=(5,5))
```

```python
edges = cv2.Canny(blur_full, threshold1=lower, threshold2=upper)
plt.imshow(edges)
```
